Write a Python Program to implement your own myreduce() function which works exactly like Python's built-in function reduce() 

```python
from functools import reduce

def myreduce(fun,l,init= None):
    iterable=iter(l)
    if init == None:
        result = next(iterable)
    else:
        result = init
    for i in iterable:
        result=fun(result,i)
    return result
```

```python
l=[2,3,4,5,6]
print(myreduce(lambda x,y:x+y,l))
print(myreduce(lambda x,y:x*y,l))
```

Write a Python program to implement your own myfilter() function which works exactly like
Python's built-in function filter()

```python
def myfilter(fun,l):
    iterable=iter(l)
    for i in iterable:
        if fun(i):
            yield i
            
l=[1,2,3,4,5,6,7,8]
for i in myfilter(lambda x:x%2==0,l):
    print(i)
```

Implement a function longestWord() that takes a list of words and returns the longest one.

```python
def longestWord(l):
    length=[len(i) for i in l]
    initial=length[0]
    for i in range(1,len(length)):
        if length[i]>initial:
            initial=length[i]
    return l[length.index(initial)]
```

```python
l=["apple","grapes","orange","pineapple"]
longestWord(l)
```

Write a Python Program(with class concepts) to find the area of the triangle using the below
formula.
area = (s*(s-a)*(s-b)*(s-c)) ** 0.5
Function to take the length of the sides of triangle from user should be defined in the parent
class and function to calculate the area should be defined in subclass.


```python
class shape:
    def getSides(self):
        self.a = int(input('Enter length of 1st side: '))
        self.b = int(input('Enter length of 2nd side: '))
        self.c = int(input('Enter length of 3rd side: '))

class triangle(shape):
    def getArea(self):
        self.getSides()
        s = (self.a + self.b + self.c) / 2
        self.area = (s*(s-self.a)*(s-self.b)*(s-self.c)) ** 0.5
        print(f'area = {self.area}')
        
t1 = triangle()
t1.getArea()
```

1.2

Write a function filter_long_words() that takes a list of words and an integer n and returns the list
of words that are longer than n.

```python
def filter_long_words(l,n):
    return [i for i in l if len(i) > n]
s=["sabrina","shawn","taylor","selena","xyz"]
filter_long_words(s,4)
```

2.2

Write a Python function which takes a character (i.e. a string of length 1) and returns True if it is
a vowel, False otherwise.


```python
def isvowel(c):
    c.lower()
    if c in ['a', 'e', 'i', 'o', 'u']:
        return True
    else:
        return False
    
print(isvowel('e'))
print(isvowel('k'))
```
